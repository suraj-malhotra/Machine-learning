```python
import numpy as np
import pandas as pd
```

```python
from sklearn.impute import SimpleImputer#missing value
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import OrdinalEncoder
```

```python
df = pd.read_csv('covid_toy.csv')
```

```python
df.shape
```

```python
df['cough'].value_counts()#gender cough city = categorical columns
```

```python
df['city'].value_counts()#gender and city nominal categorical columns
```

```python
df.isnull().sum()#missing value=fever #gender $ city=onehot encording ,fever=simple imputer ,cough=ordinal encorder
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(df.drop(columns=['has_covid']),df['has_covid'],
                                                test_size=0.2) #drop the has covid column in x_train and but not for y_train
```

```python
X_train   #'has_covid' not present #80$ train data and 20% test data
```

bad method


```python
# adding simple imputer to fever col #fill missing value in fever by simpleimputer
si = SimpleImputer() #class and si=object
X_train_fever = si.fit_transform(X_train[['fever']]) #only fever column

# also the test data
X_test_fever = si.fit_transform(X_test[['fever']])
                                 
X_train_fever   #all missing value transform into mean value
```

```python
# Ordinalencoding -> cough
oe = OrdinalEncoder(categories=[['Mild','Strong']]) # mild small value and strong high value
X_train_cough = oe.fit_transform(X_train[['cough']])

# also the test data
X_test_cough = oe.fit_transform(X_test[['cough']])

X_train_cough
```

```python
# OneHotEncoding -> gender,city #gender= 2 categories and city = 4 categories
ohe = OneHotEncoder(drop='first',sparse=False)#for multicollinearity we remove first columns from each and we get 1 columns from gender and 3 columns from city
X_train_gender_city = ohe.fit_transform(X_train[['gender','city']])

# also the test data
X_test_gender_city = ohe.fit_transform(X_test[['gender','city']])

X_train_gender_city
```

```python
# Extracting Age
X_train_age = X_train.drop(columns=['gender','fever','cough','city']).values

# also the test data
X_test_age = X_test.drop(columns=['gender','fever','cough','city']).values

X_train_age
```

```python
X_train_transformed = np.concatenate((X_train_age,X_train_fever,X_train_gender_city,X_train_cough),axis=1)
# also the test data
X_test_transformed = np.concatenate((X_test_age,X_test_fever,X_test_gender_city,X_test_cough),axis=1)

X_train_transformed
```

best method


```python
from sklearn.compose import ColumnTransformer
```

```python
transformer = ColumnTransformer(transformers=[    #transformer pass in tuple
    ('tnf1',SimpleImputer(),['fever']),
    ('tnf2',OrdinalEncoder(categories=[['Mild','Strong']]),['cough']),
    ('tnf3',OneHotEncoder(sparse=False,drop='first'),['gender','city'])
],remainder='passthrough')
```

```python
transformer.fit_transform(X_train)
```
