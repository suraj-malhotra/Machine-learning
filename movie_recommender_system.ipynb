```python
import ast
```

```python
def convert(obj):
  L=[]
  for i in ast.literal_eval(obj):
    L.append(i['name'])
  return L  
```

```python
movies['genres']=movies['genres'].apply(convert)
```

```python
movies['keywords']=movies['keywords'].apply(convert)
```

```python
movies.head(2)
```

```python
def cast1(obj):
  L=[]
  count=0
  for i in ast.literal_eval(obj):
    if count<3:
       L.append(i['name'])
       count+=1
    else:
       break
  return L  
```

```python
movies['cast']=movies['cast'].apply(cast1)
```

```python
movies.head(2)
```

```python
def director1(obj):
  L=[]
  for i in ast.literal_eval(obj):
    if i['job']=='Director':
       L.append(i['name'])
  return L  
```

```python
movies['crew'].apply(director1)
```

```python
movies['crew']=movies['crew'].apply(director1)
```

```python
movies['overview']=movies['overview'].apply(lambda x:x.split())
```

```python
movies.head(2)
```

```python
movies['genres'].apply(lambda x:[i.replace(" ","")for i in x])
```

```python
movies['cast'] = movies['cast'].apply(lambda x:[i.replace(" ","") for i in x])
movies['crew'] = movies['crew'].apply(lambda x:[i.replace(" ","") for i in x])
movies['genres'] = movies['genres'].apply(lambda x:[i.replace(" ","") for i in x])
movies['keywords'] = movies['keywords'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
movies.head(2)
```

```python
movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']
```

```python
new_movies=movies[['movie_id','title','tags']]
```

```python
new_movies.head(3)
```

```python
new_movies['tags']=new_movies['tags'].apply(lambda x:" ".join(x))
```

```python
new_movies.head(5)
```

```python
new_movies['tags'][0]
```

```python
new_movies['tags']=new_movies['tags'].apply(lambda x:x.lower())
```

```python
new_movies.head(3)
```

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
```

```python
def stem(text):
    y=[]

    for i in text.split():
        y.append(ps.stem(i)) 

    return " ".join(y)
```

```python
new_movies['tags']=new_movies['tags'].apply(stem)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features=5000,stop_words='english')
```

```python
vectors=cv.fit_transform(new_movies['tags']).toarray()
```

```python
vectors[0]
```

```python
cv.get_feature_names()
```

```python
#4806 vectors and each vector have 5000 number
#we have to find distance of each movie to another movie by cosine distance method,small angle means its showing max similarity.
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
cosine_similarity(vectors)
```

```python
cosine_similarity(vectors).shape
```

```python
similarity=cosine_similarity(vectors)
```

```python
similarity[0]
```

```python
def recommend(movie):
    movie_index= new_movies[new_movies['title']== movie].index[0]  
    distances=similarity[movie_index]
    movies_list=sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[1:6]

    for i in movies_list:
      print(new_movies.iloc[i[0]].title)

    return
                       
```

```python
recommend('Avatar')
```
